```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
train_url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vRTK2NvcndgPX41Czu6Ft2Ho_nE-z50BgTqdzwFW0rsJ2nvyNLe2DoIg1COzUbgw80oaRBjfy5-WtFk/pubhtml?urp=gmail_link'
response = requests.get(train_url)
soup = BeautifulSoup(response.text, 'html.parser')
s = soup.find('table', {'class':'waffle'})
tr = s.find_all('tr')
x = [0,0]
y = []
for i in tr:
  x = i.find_all('td', {'class':'s1'})
  if len(x)>1:
    y.append({
        'x': int(x[0].text.strip()),
        'y': x[1].text.strip()
    })
```

```python
df = pd.DataFrame(y)
df['y'] = pd.to_numeric(df['y'])
df.to_csv('COBtrain.csv', index=False)
print('Saved to COBtrain.csv')
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(df['x'], df['y'], color='turquoise')
plt.show()
```
